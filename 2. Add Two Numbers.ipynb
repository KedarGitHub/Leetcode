```python
# Definition for singly-linked list.

class LinkedNode:
    def __init__(self, val):
        self.val = val
        self.next = None 

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        first = l1
        first_track = l1
        second = l2
        second_track = l2
        prev=None
        first_len =0
        second_len=0

        while first_track:
            first_len+=1
            first_track=first_track.next
        
        while second_track:
            second_len+=1
            second_track=second_track.next
                      
        count=0

        if first_len>=second_len:
            head = first
            while first:
                prev = first
                if second:
                    first.val = (first.val+second.val+count)
                    quotient = first.val//10
                    first.val = first.val%10
                    if (quotient > 0):
                        count=1
                    else:
                        count=0
                else:
                    first.val = first.val + count
                    quotient = first.val//10
                    first.val = first.val%10
                    if (quotient > 0):
                        count=1
                    else:
                        count=0
                if second:
                    second = second.next
                    first= first.next
                else:
                    first= first.next
                
        else:
            head = second 
            while second:
                prev = second
                if first:
                    second.val = (first.val+second.val+count)
                    quotient = second.val//10
                    second.val = second.val%10
                    if (quotient > 0):
                        count=1
                    else:
                        count=0
                else:
                    second.val = second.val+count
                    quotient = second.val//10
                    second.val = second.val%10
                    if (quotient > 0):
                        count=1
                    else:
                        count=0
                if first:
                    first= first.next
                    second = second.next
                else:
                    second = second.next
        if count==1:
            new_node = LinkedNode(1)
            prev.next = new_node
        return head

    
```
