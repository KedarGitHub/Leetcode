```python
class MyLinkedList:

    def __init__(self):
        self.head=None

        

    def get(self, index: int) -> int:
        print(f' get Linked List : ')
        cur = self.head
        ind = 1
        while cur:
            print(cur.val)
            if index==ind:
                print(f' get Linked List : ')
                self.repr()
                print(f'cur.val {cur.val}')
                return cur.val
            ind+=1
            cur = cur.next
        return -1

    def addAtHead(self, val: int) -> None:
        obj = LinkedNode(val)
        obj.next = self.head
        self.head = obj
        print(f' addAtHead Linked List : ')
        self.repr()

    def addAtTail(self, val: int) -> None:
        cur = self.head
        obj = LinkedNode(val)
        while cur and cur.next:
            cur = cur.next   
        if cur:
            cur.next = obj
        else:
            self.addAtHead(val)
        print(f' addAtTail Linked List : ')
        self.repr()
        

    def addAtIndex(self, index: int, val: int) -> None:
        prev = None
        cur = self.head
        temp=0
        print(f'addAtIndex')
        if index==0 or cur==None:
            self.addAtHead(val)
        else:
            while cur:
                print(cur.val)
                if index==temp:
                    print(prev.val)
                    obj = LinkedNode(val)
                    prev.next = obj
                    obj.next=cur
                    print(f' addAtIndex Linked List : ')
                    self.repr()
                    break
                temp+=1
                prev=cur
                cur = cur.next
                
    def deleteAtIndex(self, index: int) -> None:
        prev = None
        cur = self.head
        ind=0
        if self.head!=None:
            if index==0:
                self.head = self.head.next
            else:
                while cur:
                    if index==ind:
                        prev.next=cur.next
                        print(f' deleteAtIndex Linked List : ')
                        self.repr()
                        break
                    ind+=1
                    prev=cur
                    cur=cur.next

    def repr(self):
            cur = self.head
            while cur:
                print(f'{cur.val}->', end=' ')
                cur=cur.next
            print()
        

class LinkedNode:
    def __init__(self, val: int) -> None:
        self.val= val
        self.next= None
```
