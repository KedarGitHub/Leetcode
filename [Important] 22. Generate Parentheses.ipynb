```python
def generate_parentheses(n):
    def backtrack(s, left, right):
        print(result)
        print(f'This is s : {s}')
        print(f'This is left {left}')
        print(f'This is right {right}')
        if len(s) == 2 * n:
            print('Append')
            result.append(s)
            return

        if left < n:
            print('Printing Left')
            backtrack(s + '(', left + 1, right)

        if right < left:
            print('Printing right')
            backtrack(s + ')', left, right + 1)

    result = []
    backtrack('', 0, 0)
    return result

# Example usage:
n = 3
combinations = generate_parentheses(n)
print(combinations)
```
